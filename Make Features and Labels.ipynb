```python
import pandas as pd
import numpy as np
import os


daily_temps_df = pd.read_csv("weather_station_temps/all_stations_temp.csv") # dataframe containing daily 
                                                                            # temperatures for weather stations

    
typhoon_df = pd.read_csv("typhoon_data/cleaned_typhoon_data.csv") # dataframe containing dates of hurricanes
                                                                  # and typhoons
typhoon_df.set_index("Date",inplace = True)

daily_temps_df.head(10)
```

```python
features = np.zeros((daily_temps_df.shape[0]-5,12*5+1))
"""
The i th column of the features array gives the value given by features_cols[i]. 
Here -5 means that value 5 days before the event, -4 means 4 days before, etc.


"""
features_cols = []
for i in range(-5 , 0):
    for x in daily_temps_df.columns[2:]:
        features_cols += [x + " " + str(i)]

features_cols += ["Date"]
print(features_cols)

features[:,-1] = daily_temps_df["date"][5:]

for i in range(0,12*5):
    features[:,i] = daily_temps_df[daily_temps_df.columns[(i)%12+2]][int(i/12):-5+int(i/12)]

print(features.shape)
features_df = pd.DataFrame(data = features,columns = features_cols)
#features.append(dailytemps_df["date"][5:].values)


features_df.head()
```

Now we add a labels column to this dataframe which will read '1' if there was a hurricane on that day or '0' if there wasn't

```python
labels = list(daily_temps_df["date"][5:])

print(typhoon_df.index)
for i in range(0,len(labels)):
    if labels[i] in typhoon_df.index:
        labels[i] = 1
    else:
        labels[i] = 0


features_df.insert(loc = 0,column = "label",value = labels)
```

```python
features_df.head()
features_df.to_csv("labels_and_features.csv")
```
